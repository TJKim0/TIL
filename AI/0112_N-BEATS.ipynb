# N-BEATS
## 시계열 분해
* N-BEATS 는 시계열 분해를 기반으로 최종값 예측
* Observed Data = Trend + Seasonal + Residual

## N-BEATS 구성


* M 개의 스택이 있지만 실제로는 3개만 사용함
    * Trend stack
    * Seasonal stack
    * Residual stack
* Trend stack 에서 제거된 Trend 값은 Seasonality stack 에 포함되지 않음. 그 다음도 마찬가지.
    


### Blocks
* Trend block 들은 입력된 데이터에서 추세를 추출하고 제거함.
* Seasonality block 들은 입력된 데이터에서 계절성을 추출하고 제거함.
* Residual block 들은 추세와 계절성을 제외한 나머지를 추출함.
* 각 block 에서 추출된 데이터는 다음 block 에 포함되지 않는다.

### Forecast
* 각 stack 에서 예측된 Trend, seasonality, residual(noise) 를 바탕으로 forecast  예측.
* 각 block 에서 뱉어낸 값을 compile 해서 stack forecast
